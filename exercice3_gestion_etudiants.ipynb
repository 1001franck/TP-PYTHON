```python
import pandas as pd
from datetime import datetime
import json
import os

# =============================================================================
# EXERCICE 3 : GESTION DES Ã‰TUDIANTS
# =============================================================================
# 
# Programme de gestion des Ã©tudiants avec calcul des moyennes et classement
# 
# FonctionnalitÃ©s :
# - Gestion de 25 Ã©tudiants de 8 nationalitÃ©s diffÃ©rentes
# - Calcul automatique des moyennes et mentions
# - Classement par ordre de mÃ©rite avec attribution des rangs
# - Recherche avancÃ©e (par matricule, nom, nationalitÃ©, etc.)
# - Statistiques de la classe et analyses de performance
# - Ajout, modification et suppression d'Ã©tudiants
# - Export des rÃ©sultats et sauvegarde des donnÃ©es
# 
# SystÃ¨me de mentions :
# < 10 : Insuffisant
# [10-12[ : Passable
# [12-14[ : Assez Bien
# [14-16[ : Bien
# [16-18[ : TrÃ¨s Bien
# [18-19.5[ : Excellent
# >= 19.5 : Honorable
# 
# CritÃ¨res d'Ã©valuation :
# - Admis : Moyenne >= 10/20
# - Redoublants : 8 <= Moyenne < 10/20
# - Exclus : Moyenne < 8/20
# 
# Menu principal (16 options) :
# 1-4 : Affichage et consultation
# 5-8 : Recherche et filtres
# 9-11 : Gestion des donnÃ©es
# 12-16 : Outils avancÃ©s (export, sauvegarde, aide, quitter)
# 
# DÃ©veloppÃ© pour l'exercice 3 de gestion des Ã©tudiants par FOTSI, ILOUMBOU ET YOMBI
# =============================================================================

def get_donnees_initiales():
    """Retourne une liste d'Ã©tudiants pour commencer."""
    return [
        {
            'matricule': 'ETU001', 
            'nom': 'KOUASSI', 
            'prenom': 'Jean', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [15, 18, 12],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU002', 
            'nom': 'KOUAME', 
            'prenom': 'Marie', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [17, 19, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU003', 
            'nom': 'KOUADIO', 
            'prenom': 'Pierre', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [14, 16, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU004', 
            'nom': 'KOUASSI', 
            'prenom': 'Sophie', 
            'age': 20, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [16, 14, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU005', 
            'nom': 'KOUADIO', 
            'prenom': 'Paul', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [8, 9, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU006', 
            'nom': 'KOUAME', 
            'prenom': 'Alice', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [19, 20, 19],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux Ã©tudiants - Cameroun
        {
            'matricule': 'ETU007', 
            'nom': 'MBALLA', 
            'prenom': 'David', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Camerounaise',
            'notes': [16, 17, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU008', 
            'nom': 'TCHOKOUANI', 
            'prenom': 'Fatou', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Camerounaise',
            'notes': [18, 19, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU009', 
            'nom': 'TAKAM', 
            'prenom': 'Franck', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Camerounaise',
            'notes': [12, 14, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux Ã©tudiants - Gabon
        {
            'matricule': 'ETU010', 
            'nom': 'ILOUMBOU', 
            'prenom': 'Junior', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [15, 16, 14],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU011', 
            'nom': 'NDONG', 
            'prenom': 'Thomas', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [9, 8, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU012', 
            'nom': 'YOMBI', 
            'prenom': 'Descartes', 
            'age': 24, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [20, 19, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux Ã©tudiants - Niger
        {
            'matricule': 'ETU013', 
            'nom': 'MAHAMADOU', 
            'prenom': 'Amina', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'NigÃ©rienne',
            'notes': [13, 15, 12],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU014', 
            'nom': 'ISSOUFOU', 
            'prenom': 'Moussa', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'NigÃ©rienne',
            'notes': [11, 9, 10],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU015', 
            'nom': 'TANDJA', 
            'prenom': 'Hassan', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'NigÃ©rienne',
            'notes': [17, 18, 16],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux Ã©tudiants - Tchad
        {
            'matricule': 'ETU016', 
            'nom': 'DEBY', 
            'prenom': 'Zara', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'Tchadienne',
            'notes': [14, 13, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU017', 
            'nom': 'KABADI', 
            'prenom': 'Idriss', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Tchadienne',
            'notes': [6, 5, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU018', 
            'nom': 'MAHAMAT', 
            'prenom': 'Fatima', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Tchadienne',
            'notes': [19, 20, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # 4 Ã©tudiants supplÃ©mentaires
        {
            'matricule': 'ETU019', 
            'nom': 'KONATE', 
            'prenom': 'Moussa', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'Malienne',
            'notes': [14, 16, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU020', 
            'nom': 'DIALLO', 
            'prenom': 'Aissatou', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'GuinÃ©enne',
            'notes': [8, 7, 6],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU021', 
            'nom': 'TOURE', 
            'prenom': 'SÃ©kou', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Malienne',
            'notes': [18, 19, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU022', 
            'nom': 'CAMARA', 
            'prenom': 'Mariama', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'GuinÃ©enne',
            'notes': [20, 19, 20],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # 3 Ã©tudiants supplÃ©mentaires
        {
            'matricule': 'ETU023', 
            'nom': 'SALL', 
            'prenom': 'Macky', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'SÃ©nÃ©galaise',
            'notes': [16, 15, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU024', 
            'nom': 'WADE', 
            'prenom': 'Aminata', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'SÃ©nÃ©galaise',
            'notes': [12, 11, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU025', 
            'nom': 'DIOP', 
            'prenom': 'Cheikh', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'SÃ©nÃ©galaise',
            'notes': [19, 20, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        }
    ]

def calculer_moyennes(liste_etudiants):
    """Calcule la moyenne de chaque Ã©tudiant."""
    for etudiant in liste_etudiants:
        if len(etudiant['notes']) == 3:
            etudiant['moyenne'] = sum(etudiant['notes']) / 3
        else:
            etudiant['moyenne'] = 0
    return liste_etudiants

def determiner_mentions(liste_etudiants):
    """DÃ©termine la mention de chaque Ã©tudiant."""
    for etudiant in liste_etudiants:
        moyenne = etudiant['moyenne']
        if moyenne < 10:
            etudiant['mention'] = 'Insuffisant'
        elif moyenne < 12:
            etudiant['mention'] = 'Passable'
        elif moyenne < 14:
            etudiant['mention'] = 'Assez Bien'
        elif moyenne < 16:
            etudiant['mention'] = 'Bien'
        elif moyenne < 18:
            etudiant['mention'] = 'TrÃ¨s Bien'
        elif moyenne < 19.5:
            etudiant['mention'] = 'Excellent'
        else:
            etudiant['mention'] = 'Honorable'
    return liste_etudiants

def classer_par_merite(liste_etudiants):
    """Classe les Ã©tudiants par ordre de mÃ©rite et affecte le rang."""
    # Trier par moyenne dÃ©croissante
    liste_triee = sorted(liste_etudiants, key=lambda x: x['moyenne'], reverse=True)
    
    # Affecter le rang
    for i, etudiant in enumerate(liste_triee, 1):
        etudiant['rang'] = i
    
    return liste_triee

def afficher_informations_etudiants(liste_etudiants):
    """Affiche les informations de base des Ã©tudiants."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # En-tÃªtes des colonnes
    headers = ['Matricule', 'NOM', 'PrÃ©nom', 'Age', 'Sexe', 'NationalitÃ©', 'Notes']
    
    # PrÃ©parer les donnÃ©es pour l'affichage
    data = []
    for etudiant in liste_etudiants:
        notes_str = f"{etudiant['notes'][0]}, {etudiant['notes'][1]}, {etudiant['notes'][2]}"
        
        data.append([
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            str(etudiant['age']),
            etudiant['sexe'],
            etudiant['nationalite'],
            notes_str
        ])
    
    # Calculer la largeur de chaque colonne
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour crÃ©er une ligne de sÃ©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-tÃªtes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # DonnÃ©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def afficher_resultats_etudiants(liste_etudiants):
    """Affiche les rÃ©sultats des Ã©tudiants (moyennes, mentions, classement)."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # Trier les Ã©tudiants par rang (ordre de mÃ©rite)
    etudiants_tries = sorted(liste_etudiants, key=lambda x: x['rang'])
    
    # En-tÃªtes des colonnes
    headers = ['Rang', 'Matricule', 'NOM', 'PrÃ©nom', 'Moyenne', 'Mention', 'Statut']
    
    # PrÃ©parer les donnÃ©es pour l'affichage
    data = []
    for etudiant in etudiants_tries:
        # DÃ©terminer le statut
        moyenne = etudiant['moyenne']
        if moyenne >= 10:
            statut = "ðŸŸ¢ Admis"
        elif moyenne >= 8:
            statut = "ðŸ”´ Redouble"
        else:
            statut = "ðŸ”´ Exclu"
        
        # DÃ©terminer l'emoji selon la mention
        mention = etudiant['mention']
        if mention in ['Honorable', 'Excellent']:
            emoji = "ðŸŸ¢"
        elif mention in ['TrÃ¨s Bien', 'Bien']:
            emoji = "ðŸŸ¡"
        elif mention in ['Assez Bien', 'Passable']:
            emoji = "ðŸŸ "
        else:
            emoji = "ðŸ”´"
        
        data.append([
            str(etudiant['rang']),
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            f"{etudiant['moyenne']:.2f}",
            f"{emoji} {etudiant['mention']}",
            statut
        ])
    
    # Calculer la largeur de chaque colonne avec une largeur minimale pour Statut
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        
        # Largeur minimale pour la colonne Statut
        if header == 'Statut':
            max_width = max(max_width, 12)  # Au moins 12 caractÃ¨res pour "ðŸ”´ Redouble"
        
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour crÃ©er une ligne de sÃ©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-tÃªtes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # DonnÃ©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def afficher_admis(liste_etudiants):
    """Affiche les Ã©tudiants admis (moyenne >= 10)."""
    admis = [etudiant for etudiant in liste_etudiants if etudiant['moyenne'] >= 10]
    if admis:
        print(f"\n=== Ã‰TUDIANTS ADMIS ({len(admis)} Ã©tudiants) ===")
        afficher_informations_etudiants(admis)
    else:
        print("\nAucun Ã©tudiant admis.")

def afficher_redoublants(liste_etudiants):
    """Affiche les redoublants (moyenne entre 8 et 10)."""
    redoublants = [etudiant for etudiant in liste_etudiants if 8 <= etudiant['moyenne'] < 10]
    if redoublants:
        print(f"\n=== REDOUBLANTS ({len(redoublants)} Ã©tudiants) ===")
        afficher_informations_etudiants(redoublants)
    else:
        print("\nAucun redoublant.")

def afficher_exclus(liste_etudiants):
    """Affiche les exclus (moyenne < 8)."""
    exclus = [etudiant for etudiant in liste_etudiants if etudiant['moyenne'] < 8]
    if exclus:
        print(f"\n=== EXCLUS ({len(exclus)} Ã©tudiants) ===")
        afficher_informations_etudiants(exclus)
    else:
        print("\nAucun Ã©tudiant exclu.")

def afficher_etudiants_alphabÃ©tique(liste_etudiants):
    """Affiche les Ã©tudiants par ordre alphabÃ©tique (nom, puis prÃ©nom)."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # Trier les Ã©tudiants par ordre alphabÃ©tique (nom, puis prÃ©nom)
    etudiants_tries = sorted(liste_etudiants, key=lambda x: (x['nom'], x['prenom']))
    
    # En-tÃªtes des colonnes
    headers = ['NÂ°', 'Matricule', 'NOM', 'PrÃ©nom', 'Age', 'Sexe', 'NationalitÃ©', 'Notes']
    
    # PrÃ©parer les donnÃ©es pour l'affichage
    data = []
    for i, etudiant in enumerate(etudiants_tries, 1):
        notes_str = f"{etudiant['notes'][0]}, {etudiant['notes'][1]}, {etudiant['notes'][2]}"
        
        data.append([
            str(i),
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            str(etudiant['age']),
            etudiant['sexe'],
            etudiant['nationalite'],
            notes_str
        ])
    
    # Calculer la largeur de chaque colonne
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour crÃ©er une ligne de sÃ©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-tÃªtes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # DonnÃ©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def calculer_statistiques(liste_etudiants):
    """Calcule les statistiques avancÃ©es de la classe."""
    if not liste_etudiants:
        print("\nAucun Ã©tudiant pour calculer les statistiques.")
        return
    
    # Statistiques de base
    nb_total = len(liste_etudiants)
    moyennes = [etudiant['moyenne'] for etudiant in liste_etudiants]
    moyenne_generale = sum(moyennes) / nb_total
    note_min = min(moyennes)
    note_max = max(moyennes)
    
    # Ã‰cart-type
    variance = sum((m - moyenne_generale) ** 2 for m in moyennes) / nb_total
    ecart_type = variance ** 0.5
    
    # RÃ©partition par nationalitÃ©
    nationalites = {}
    for etudiant in liste_etudiants:
        nat = etudiant['nationalite']
        nationalites[nat] = nationalites.get(nat, 0) + 1
    
    # RÃ©partition par sexe
    hommes = len([e for e in liste_etudiants if e['sexe'] == 'M'])
    femmes = len([e for e in liste_etudiants if e['sexe'] == 'F'])
    
    # RÃ©partition par mention
    mentions = {}
    for etudiant in liste_etudiants:
        mention = etudiant['mention']
        mentions[mention] = mentions.get(mention, 0) + 1
    
    # RÃ©partition par statut
    admis = len([e for e in liste_etudiants if e['moyenne'] >= 10])
    redoublants = len([e for e in liste_etudiants if 8 <= e['moyenne'] < 10])
    exclus = len([e for e in liste_etudiants if e['moyenne'] < 8])
    
    # Affichage des statistiques
    print("\n" + "="*60)
    print("ðŸ“Š STATISTIQUES DE LA CLASSE")
    print("="*60)
    
    print(f"\nðŸ‘¥ Effectif total : {nb_total} Ã©tudiants")
    print(f"ðŸ“ˆ Moyenne gÃ©nÃ©rale : {moyenne_generale:.2f}/20")
    print(f"ðŸ“Š Note la plus basse : {note_min:.2f}/20")
    print(f"ðŸ“Š Note la plus haute : {note_max:.2f}/20")
    print(f"ðŸ“ Ã‰cart-type : {ecart_type:.2f}")
    
    print(f"\nðŸ‘¨â€ðŸŽ“ RÃ©partition par sexe :")
    print(f"   Hommes : {hommes} ({hommes/nb_total*100:.1f}%)")
    print(f"   Femmes : {femmes} ({femmes/nb_total*100:.1f}%)")
    
    print(f"\nðŸŒ RÃ©partition par nationalitÃ© :")
    for nat, nb in sorted(nationalites.items()):
        print(f"   {nat} : {nb} Ã©tudiant(s) ({nb/nb_total*100:.1f}%)")
    
    print(f"\nðŸ† RÃ©partition par mention :")
    for mention, nb in sorted(mentions.items()):
        print(f"   {mention} : {nb} Ã©tudiant(s) ({nb/nb_total*100:.1f}%)")
    
    print(f"\nðŸ“‹ RÃ©partition par statut :")
    print(f"   ðŸŸ¢ Admis : {admis} Ã©tudiant(s) ({admis/nb_total*100:.1f}%)")
    print(f"   ðŸ”´ Redoublants : {redoublants} Ã©tudiant(s) ({redoublants/nb_total*100:.1f}%)")
    print(f"   ðŸ”´ Exclus : {exclus} Ã©tudiant(s) ({exclus/nb_total*100:.1f}%)")
    
    print("="*60)

def rechercher_etudiant_complet(liste_etudiants):
    """Recherche complÃ¨te d'Ã©tudiants avec plusieurs options."""
    print("\n--- RECHERCHE D'Ã‰TUDIANTS ---")
    print("1. Rechercher par matricule")
    print("2. Rechercher par nom/prÃ©nom")
    print("3. Rechercher par nationalitÃ©")
    print("4. Rechercher par mention")
    print("5. Rechercher par statut")
    print("6. Rechercher par tranche d'Ã¢ge")
    print("7. Retour au menu principal")
    
    choix_recherche = input("Votre choix : ")
    
    if choix_recherche == '1':
        matricule = input("Entrez le matricule de l'Ã©tudiant Ã  rechercher : ")
        etudiant_trouve = rechercher_etudiant(liste_etudiants, matricule)
        if etudiant_trouve:
            print(f"\n=== DÃ‰TAILS DE L'Ã‰TUDIANT ===")
            print(f"Matricule : {etudiant_trouve['matricule']}")
            print(f"Nom : {etudiant_trouve['nom']}")
            print(f"PrÃ©nom : {etudiant_trouve['prenom']}")
            print(f"Age : {etudiant_trouve['age']} ans")
            print(f"Sexe : {etudiant_trouve['sexe']}")
            print(f"NationalitÃ© : {etudiant_trouve['nationalite']}")
            print(f"Notes : {etudiant_trouve['notes'][0]}, {etudiant_trouve['notes'][1]}, {etudiant_trouve['notes'][2]}")
            print(f"Moyenne : {etudiant_trouve['moyenne']:.2f}")
            print(f"Mention : {etudiant_trouve['mention']}")
            print(f"Rang : {etudiant_trouve['rang']}")
        else:
            print("âŒ Aucun Ã©tudiant trouvÃ© avec ce matricule.")
    
    elif choix_recherche == '2':
        terme = input("Entrez le nom ou prÃ©nom Ã  rechercher : ").upper()
        resultats = []
        for etudiant in liste_etudiants:
            if terme in etudiant['nom'].upper() or terme in etudiant['prenom'].upper():
                resultats.append(etudiant)
        
        if resultats:
            print(f"\nâœ… {len(resultats)} Ã©tudiant(s) trouvÃ©(s) :")
            afficher_informations_etudiants(resultats)
        else:
            print("âŒ Aucun Ã©tudiant trouvÃ©.")
    
    elif choix_recherche == '3':
        print("\nNationalitÃ©s disponibles :")
        nationalites = list(set(etudiant['nationalite'] for etudiant in liste_etudiants))
        for i, nat in enumerate(nationalites, 1):
            print(f"{i}. {nat}")
        
        try:
            choix = int(input("Choisissez le numÃ©ro de la nationalitÃ© : ")) - 1
            if 0 <= choix < len(nationalites):
                nationalite = nationalites[choix]
                resultats = [e for e in liste_etudiants if e['nationalite'] == nationalite]
                print(f"\nâœ… {len(resultats)} Ã©tudiant(s) {nationalite}(s) :")
                afficher_informations_etudiants(resultats)
            else:
                print("âŒ Choix invalide.")
        except ValueError:
            print("âŒ Veuillez entrer un numÃ©ro valide.")
    
    elif choix_recherche == '4':
        print("\nMentions disponibles :")
        mentions = ['Insuffisant', 'Passable', 'Assez Bien', 'Bien', 'TrÃ¨s Bien', 'Excellent', 'Honorable']
        for i, mention in enumerate(mentions, 1):
            print(f"{i}. {mention}")
        
        try:
            choix = int(input("Choisissez le numÃ©ro de la mention : ")) - 1
            if 0 <= choix < len(mentions):
                mention = mentions[choix]
                resultats = [e for e in liste_etudiants if e['mention'] == mention]
                if resultats:
                    print(f"\nâœ… {len(resultats)} Ã©tudiant(s) avec mention '{mention}' :")
                    afficher_informations_etudiants(resultats)
                else:
                    print(f"âŒ Aucun Ã©tudiant avec la mention '{mention}'.")
            else:
                print("âŒ Choix invalide.")
        except ValueError:
            print("âŒ Veuillez entrer un numÃ©ro valide.")
    
    elif choix_recherche == '5':
        print("\nStatuts disponibles :")
        print("1. Admis (moyenne >= 10)")
        print("2. Redoublants (8 <= moyenne < 10)")
        print("3. Exclus (moyenne < 8)")
        
        choix = input("Choisissez le statut (1-3) : ")
        if choix == '1':
            resultats = [e for e in liste_etudiants if e['moyenne'] >= 10]
            print(f"\nâœ… {len(resultats)} Ã©tudiant(s) admis :")
        elif choix == '2':
            resultats = [e for e in liste_etudiants if 8 <= e['moyenne'] < 10]
            print(f"\nâœ… {len(resultats)} redoublant(s) :")
        elif choix == '3':
            resultats = [e for e in liste_etudiants if e['moyenne'] < 8]
            print(f"\nâœ… {len(resultats)} Ã©tudiant(s) exclu(s) :")
        else:
            print("âŒ Choix invalide.")
            return
        
        if resultats:
            afficher_informations_etudiants(resultats)
        else:
            print("âŒ Aucun Ã©tudiant trouvÃ©.")
    
    elif choix_recherche == '6':
        try:
            age_min = int(input("Ã‚ge minimum : "))
            age_max = int(input("Ã‚ge maximum : "))
            resultats = [e for e in liste_etudiants if age_min <= e['age'] <= age_max]
            
            if resultats:
                print(f"\nâœ… {len(resultats)} Ã©tudiant(s) entre {age_min} et {age_max} ans :")
                afficher_informations_etudiants(resultats)
            else:
                print("âŒ Aucun Ã©tudiant trouvÃ© dans cette tranche d'Ã¢ge.")
        except ValueError:
            print("âŒ Veuillez entrer des Ã¢ges valides.")
    
    elif choix_recherche == '7':
        return
    
    else:
        print("âŒ Choix invalide.")

def valider_donnees_etudiant(matricule, nom, prenom, age, sexe, nationalite, notes):
    """Valide les donnÃ©es d'un Ã©tudiant avant ajout/modification."""
    erreurs = []
    
    # Validation du matricule
    if not matricule or len(matricule.strip()) < 3:
        erreurs.append("Le matricule doit contenir au moins 3 caractÃ¨res")
    
    # Validation du nom et prÃ©nom
    if not nom or len(nom.strip()) < 2:
        erreurs.append("Le nom doit contenir au moins 2 caractÃ¨res")
    if not prenom or len(prenom.strip()) < 2:
        erreurs.append("Le prÃ©nom doit contenir au moins 2 caractÃ¨res")
    
    # Validation de l'Ã¢ge
    if not isinstance(age, int) or age < 15 or age > 35:
        erreurs.append("L'Ã¢ge doit Ãªtre entre 15 et 35 ans")
    
    # Validation du sexe
    if sexe not in ['M', 'F']:
        erreurs.append("Le sexe doit Ãªtre 'M' ou 'F'")
    
    # Validation de la nationalitÃ©
    if not nationalite or len(nationalite.strip()) < 3:
        erreurs.append("La nationalitÃ© doit contenir au moins 3 caractÃ¨res")
    
    # Validation des notes
    if len(notes) != 3:
        erreurs.append("Il faut exactement 3 notes")
    else:
        for i, note in enumerate(notes, 1):
            if not isinstance(note, (int, float)) or note < 0 or note > 20:
                erreurs.append(f"La note {i} doit Ãªtre entre 0 et 20")
    
    return erreurs

def exporter_resultats(liste_etudiants):
    """Exporte les rÃ©sultats dans un fichier texte."""
    try:
        nom_fichier = f"resultats_etudiants_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
        
        with open(nom_fichier, 'w', encoding='utf-8') as f:
            f.write("="*80 + "\n")
            f.write("RAPPORT DES RÃ‰SULTATS - GESTION DES Ã‰TUDIANTS\n")
            f.write("="*80 + "\n\n")
            
            # Informations gÃ©nÃ©rales
            f.write(f"Date de gÃ©nÃ©ration : {datetime.now().strftime('%d/%m/%Y Ã  %H:%M')}\n")
            f.write(f"Nombre total d'Ã©tudiants : {len(liste_etudiants)}\n\n")
            
            # Statistiques
            moyennes = [e['moyenne'] for e in liste_etudiants]
            moyenne_generale = sum(moyennes) / len(moyennes)
            f.write(f"Moyenne gÃ©nÃ©rale de la classe : {moyenne_generale:.2f}/20\n\n")
            
            # Classement complet
            f.write("CLASSEMENT PAR ORDRE DE MÃ‰RITE\n")
            f.write("-"*80 + "\n")
            f.write(f"{'Rang':<5} {'Matricule':<10} {'Nom':<15} {'PrÃ©nom':<15} {'Moyenne':<8} {'Mention':<12} {'Statut':<10}\n")
            f.write("-"*80 + "\n")
            
            etudiants_tries = sorted(liste_etudiants, key=lambda x: x['rang'])
            for etudiant in etudiants_tries:
                statut = "Admis" if etudiant['moyenne'] >= 10 else "Redouble" if etudiant['moyenne'] >= 8 else "Exclu"
                f.write(f"{etudiant['rang']:<5} {etudiant['matricule']:<10} {etudiant['nom']:<15} {etudiant['prenom']:<15} {etudiant['moyenne']:<8.2f} {etudiant['mention']:<12} {statut:<10}\n")
            
            f.write("\n" + "="*80 + "\n")
            f.write("FIN DU RAPPORT\n")
            f.write("="*80 + "\n")
        
        print(f"âœ… Rapport exportÃ© avec succÃ¨s : {nom_fichier}")
        return True
        
    except Exception as e:
        print(f"âŒ Erreur lors de l'export : {e}")
        return False

def sauvegarder_donnees(liste_etudiants):
    """Sauvegarde les donnÃ©es dans un fichier JSON."""
    try:
        nom_fichier = f"backup_etudiants_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
        
        # PrÃ©parer les donnÃ©es pour la sauvegarde
        donnees_sauvegarde = []
        for etudiant in liste_etudiants:
            donnees_sauvegarde.append({
                'matricule': etudiant['matricule'],
                'nom': etudiant['nom'],
                'prenom': etudiant['prenom'],
                'age': etudiant['age'],
                'sexe': etudiant['sexe'],
                'nationalite': etudiant['nationalite'],
                'notes': etudiant['notes']
            })
        
        with open(nom_fichier, 'w', encoding='utf-8') as f:
            json.dump(donnees_sauvegarde, f, indent=2, ensure_ascii=False)
        
        print(f"âœ… Sauvegarde crÃ©Ã©e avec succÃ¨s : {nom_fichier}")
        return True
        
    except Exception as e:
        print(f"âŒ Erreur lors de la sauvegarde : {e}")
        return False

def restaurer_donnees():
    """Restaure les donnÃ©es depuis un fichier JSON."""
    try:
        # Lister les fichiers de sauvegarde disponibles
        fichiers_backup = [f for f in os.listdir('.') if f.startswith('backup_etudiants_') and f.endswith('.json')]
        
        if not fichiers_backup:
            print("âŒ Aucun fichier de sauvegarde trouvÃ©.")
            return None
        
        print("\nðŸ“ Fichiers de sauvegarde disponibles :")
        for i, fichier in enumerate(sorted(fichiers_backup, reverse=True), 1):
            print(f"   {i}. {fichier}")
        
        choix = input("\nChoisissez le numÃ©ro du fichier Ã  restaurer : ")
        try:
            index = int(choix) - 1
            if 0 <= index < len(fichiers_backup):
                fichier_choisi = sorted(fichiers_backup, reverse=True)[index]
                
                with open(fichier_choisi, 'r', encoding='utf-8') as f:
                    donnees = json.load(f)
                
                # Convertir les donnÃ©es en format Ã©tudiant
                etudiants_restaures = []
                for donnee in donnees:
                    etudiant = {
                        'matricule': donnee['matricule'],
                        'nom': donnee['nom'],
                        'prenom': donnee['prenom'],
                        'age': donnee['age'],
                        'sexe': donnee['sexe'],
                        'nationalite': donnee['nationalite'],
                        'notes': donnee['notes'],
                        'moyenne': 0,
                        'mention': '',
                        'rang': 0
                    }
                    etudiants_restaures.append(etudiant)
                
                # Recalculer tout
                calculer_moyennes(etudiants_restaures)
                determiner_mentions(etudiants_restaures)
                classer_par_merite(etudiants_restaures)
                
                print(f"âœ… DonnÃ©es restaurÃ©es depuis {fichier_choisi}")
                print(f"ðŸ“Š {len(etudiants_restaures)} Ã©tudiants chargÃ©s")
                return etudiants_restaures
            else:
                print("âŒ Choix invalide.")
                return None
        except ValueError:
            print("âŒ Veuillez entrer un numÃ©ro valide.")
            return None
            
    except Exception as e:
        print(f"âŒ Erreur lors de la restauration : {e}")
        return None

def analyser_performances(liste_etudiants):
    """Analyse dÃ©taillÃ©e des performances de la classe."""
    if not liste_etudiants:
        print("âŒ Aucun Ã©tudiant pour analyser.")
        return
    
    print("\n" + "="*60)
    print("ðŸ“ˆ ANALYSE DÃ‰TAILLÃ‰E DES PERFORMANCES")
    print("="*60)
    
    # Statistiques de base
    nb_total = len(liste_etudiants)
    moyennes = [e['moyenne'] for e in liste_etudiants]
    
    # Meilleurs et moins bons
    meilleur = max(liste_etudiants, key=lambda x: x['moyenne'])
    moins_bon = min(liste_etudiants, key=lambda x: x['moyenne'])
    
    # RÃ©partition par tranches de notes
    excellent = len([e for e in liste_etudiants if e['moyenne'] >= 16])
    tres_bien = len([e for e in liste_etudiants if 14 <= e['moyenne'] < 16])
    bien = len([e for e in liste_etudiants if 12 <= e['moyenne'] < 14])
    passable = len([e for e in liste_etudiants if 10 <= e['moyenne'] < 12])
    insuffisant = len([e for e in liste_etudiants if e['moyenne'] < 10])
    
    print(f"\nðŸ† MEILLEURS Ã‰LÃˆVES :")
    print(f"   â€¢ 1er : {meilleur['prenom']} {meilleur['nom']} - {meilleur['moyenne']:.2f}/20 ({meilleur['mention']})")
    
    # Trouver le 2Ã¨me et 3Ã¨me
    etudiants_tries = sorted(liste_etudiants, key=lambda x: x['moyenne'], reverse=True)
    if len(etudiants_tries) >= 2:
        deuxieme = etudiants_tries[1]
        print(f"   â€¢ 2Ã¨me : {deuxieme['prenom']} {deuxieme['nom']} - {deuxieme['moyenne']:.2f}/20 ({deuxieme['mention']})")
    if len(etudiants_tries) >= 3:
        troisieme = etudiants_tries[2]
        print(f"   â€¢ 3Ã¨me : {troisieme['prenom']} {troisieme['nom']} - {troisieme['moyenne']:.2f}/20 ({troisieme['mention']})")
    
    print(f"\nðŸ“‰ Ã‰LÃˆVES EN DIFFICULTÃ‰ :")
    print(f"   â€¢ Plus faible : {moins_bon['prenom']} {moins_bon['nom']} - {moins_bon['moyenne']:.2f}/20 ({moins_bon['mention']})")
    
    print(f"\nðŸ“Š RÃ‰PARTITION PAR NIVEAU :")
    print(f"   â€¢ Excellent (â‰¥16) : {excellent} Ã©tudiant(s) ({excellent/nb_total*100:.1f}%)")
    print(f"   â€¢ TrÃ¨s Bien (14-16) : {tres_bien} Ã©tudiant(s) ({tres_bien/nb_total*100:.1f}%)")
    print(f"   â€¢ Bien (12-14) : {bien} Ã©tudiant(s) ({bien/nb_total*100:.1f}%)")
    print(f"   â€¢ Passable (10-12) : {passable} Ã©tudiant(s) ({passable/nb_total*100:.1f}%)")
    print(f"   â€¢ Insuffisant (<10) : {insuffisant} Ã©tudiant(s) ({insuffisant/nb_total*100:.1f}%)")
    
    # Analyse par nationalitÃ©
    print(f"\nðŸŒ PERFORMANCES PAR NATIONALITÃ‰ :")
    nationalites = {}
    for etudiant in liste_etudiants:
        nat = etudiant['nationalite']
        if nat not in nationalites:
            nationalites[nat] = {'total': 0, 'moyennes': []}
        nationalites[nat]['total'] += 1
        nationalites[nat]['moyennes'].append(etudiant['moyenne'])
    
    for nat, data in sorted(nationalites.items()):
        moyenne_nat = sum(data['moyennes']) / len(data['moyennes'])
        print(f"   â€¢ {nat} : {data['total']} Ã©tudiant(s) - Moyenne : {moyenne_nat:.2f}/20")
    
    print("="*60)

def afficher_aide():
    """Affiche l'aide et les instructions d'utilisation."""
    print("\n" + "="*60)
    print("â“ AIDE ET INSTRUCTIONS")
    print("="*60)
    
    print("\nðŸ“‹ FONCTIONNALITÃ‰S PRINCIPALES :")
    print("   â€¢ Gestion complÃ¨te des Ã©tudiants (25 Ã©tudiants de 8 nationalitÃ©s)")
    print("   â€¢ Calcul automatique des moyennes et mentions")
    print("   â€¢ Classement par ordre de mÃ©rite")
    print("   â€¢ Recherche avancÃ©e multi-critÃ¨res")
    print("   â€¢ Statistiques dÃ©taillÃ©es de la classe")
    
    print("\nðŸŽ¯ SYSTÃˆME DE MENTIONS :")
    print("   â€¢ < 10 : Insuffisant")
    print("   â€¢ [10-12[ : Passable")
    print("   â€¢ [12-14[ : Assez Bien")
    print("   â€¢ [14-16[ : Bien")
    print("   â€¢ [16-18[ : TrÃ¨s Bien")
    print("   â€¢ [18-19.5[ : Excellent")
    print("   â€¢ â‰¥ 19.5 : Honorable")
    
    print("\nðŸ“Š CRITÃˆRES D'Ã‰VALUATION :")
    print("   â€¢ Admis : Moyenne â‰¥ 10/20")
    print("   â€¢ Redoublants : 8 â‰¤ Moyenne < 10/20")
    print("   â€¢ Exclus : Moyenne < 8/20")
    
    print("\nðŸ’¡ CONSEILS D'UTILISATION :")
    print("   â€¢ Utilisez les options 1-3 pour consulter les donnÃ©es")
    print("   â€¢ L'option 4 permet une recherche complÃ¨te")
    print("   â€¢ Sauvegardez rÃ©guliÃ¨rement vos donnÃ©es (option 12)")
    print("   â€¢ Exportez les rÃ©sultats pour les partager (option 11)")
    
    print("\nðŸ”§ VALIDATION DES DONNÃ‰ES :")
    print("   â€¢ Matricule : minimum 3 caractÃ¨res, unique")
    print("   â€¢ Nom/PrÃ©nom : minimum 2 caractÃ¨res")
    print("   â€¢ Ã‚ge : entre 15 et 35 ans")
    print("   â€¢ Notes : entre 0 et 20 sur 20")
    
    print("="*60)

def afficher_menu_principal():
    """Affiche le menu principal avec une meilleure organisation."""
    print("\n" + "="*60)
    print("ðŸŽ“ SYSTÃˆME DE GESTION DES Ã‰TUDIANTS")
    print("="*60)
    print("\nðŸ“‹ AFFICHAGE ET CONSULTATION")
    print("  1. Liste des Ã©tudiants (ordre alphabÃ©tique)")
    print("  2. Classement par ordre de mÃ©rite")
    print("  3. Statistiques de la classe")
    print("  4. Analyse dÃ©taillÃ©e des performances")
    print("\nðŸ” RECHERCHE ET FILTRES")
    print("  5. Rechercher un Ã©tudiant")
    print("  6. Afficher les admis")
    print("  7. Afficher les redoublants")
    print("  8. Afficher les exclus")
    print("\nâœï¸ GESTION DES DONNÃ‰ES")
    print("  9. Ajouter un Ã©tudiant")
    print("  10. Modifier un Ã©tudiant")
    print("  11. Supprimer un Ã©tudiant")
    print("\nðŸ“Š OUTILS AVANCÃ‰S")
    print("  12. Exporter les rÃ©sultats")
    print("  13. Sauvegarder les donnÃ©es")
    print("  14. Restaurer les donnÃ©es")
    print("  15. Aide et instructions")
    print("  16. Quitter")
    print("\n" + "="*60)

def menu():
    """Affiche le menu principal et retourne le choix de l'utilisateur."""
    afficher_menu_principal()
    return input("Votre choix : ")

def rechercher_etudiant(liste_etudiants, matricule):
    """Recherche un Ã©tudiant par son matricule."""
    for etudiant in liste_etudiants:
        if etudiant['matricule'].lower() == matricule.lower():
            return etudiant
    return None

def ajouter_etudiant(liste_etudiants):
    """Ajoute un nouvel Ã©tudiant Ã  la liste avec validation."""
    print("\n" + "="*50)
    print("âž• AJOUT D'UN NOUVEL Ã‰TUDIANT")
    print("="*50)
    
    try:
        # Saisie des donnÃ©es
        matricule = input("Matricule : ").strip()
        nom = input("Nom : ").strip().upper()
        prenom = input("PrÃ©nom : ").strip().title()
        age = int(input("Age : "))
        sexe = input("Sexe (M/F) : ").strip().upper()
        nationalite = input("NationalitÃ© : ").strip().title()
        
        print("\nðŸ“ Saisie des 3 notes (sur 20) :")
        notes = []
        for i in range(3):
            while True:
                try:
                    note = float(input(f"Note {i+1} : "))
                    if 0 <= note <= 20:
                        notes.append(note)
                        break
                    else:
                        print("âŒ La note doit Ãªtre entre 0 et 20.")
                except ValueError:
                    print("âŒ Veuillez entrer un nombre valide.")

        # Validation des donnÃ©es
        erreurs = valider_donnees_etudiant(matricule, nom, prenom, age, sexe, nationalite, notes)
        
        if erreurs:
            print("\nâŒ Erreurs de validation :")
            for erreur in erreurs:
                print(f"   â€¢ {erreur}")
            return

        # VÃ©rifier si le matricule existe dÃ©jÃ 
        if rechercher_etudiant(liste_etudiants, matricule):
            print(f"âŒ Le matricule {matricule} existe dÃ©jÃ .")
            return

        # CrÃ©er le nouvel Ã©tudiant
        nouvel_etudiant = {
            'matricule': matricule,
            'nom': nom,
            'prenom': prenom,
            'age': age,
            'sexe': sexe,
            'nationalite': nationalite,
            'notes': notes,
            'moyenne': 0,
            'mention': '',
            'rang': 0
        }
        
        liste_etudiants.append(nouvel_etudiant)
        
        # Recalculer tout
        calculer_moyennes(liste_etudiants)
        determiner_mentions(liste_etudiants)
        classer_par_merite(liste_etudiants)
        
        print(f"\nâœ… L'Ã©tudiant {prenom} {nom} a Ã©tÃ© ajoutÃ© avec succÃ¨s !")
        print(f"ðŸ“Š Moyenne : {nouvel_etudiant['moyenne']:.2f}/20")
        print(f"ðŸ† Mention : {nouvel_etudiant['mention']}")
        print(f"ðŸ“ˆ Rang : {nouvel_etudiant['rang']}")

    except ValueError:
        print("âŒ Erreur : L'Ã¢ge doit Ãªtre un nombre entier.")
    except Exception as e:
        print(f"âŒ Une erreur inattendue est survenue : {e}")

def main():
    etudiants = get_donnees_initiales()
    
    # Initialisation : calculer moyennes, mentions et classement
    calculer_moyennes(etudiants)
    determiner_mentions(etudiants)
    classer_par_merite(etudiants)

    print("ðŸŽ“ Bienvenue dans le systÃ¨me de gestion des Ã©tudiants !")
    print(f"ðŸ“š {len(etudiants)} Ã©tudiants chargÃ©s avec succÃ¨s.")

    while True:
        choix = menu()
        
        if choix == '1':
            afficher_etudiants_alphabÃ©tique(etudiants)
        
        elif choix == '2':
            afficher_resultats_etudiants(etudiants)
        
        elif choix == '3':
            calculer_statistiques(etudiants)
        
        elif choix == '4':
            analyser_performances(etudiants)
        
        elif choix == '5':
            rechercher_etudiant_complet(etudiants)
        
        elif choix == '6':
            afficher_admis(etudiants)
        
        elif choix == '7':
            afficher_redoublants(etudiants)
        
        elif choix == '8':
            afficher_exclus(etudiants)
        
        elif choix == '9':
            ajouter_etudiant(etudiants)
        
        elif choix == '10':
            matricule = input("Entrez le matricule de l'Ã©tudiant Ã  modifier : ")
            etudiant = rechercher_etudiant(etudiants, matricule)
            if not etudiant:
                print("âŒ Aucun Ã©tudiant trouvÃ© avec ce matricule.")
                continue
            
            print(f"\nâœï¸ Modification de {etudiant['prenom']} {etudiant['nom']}")
            print("Laissez vide pour ne pas modifier.")
            
            try:
                nouveau_nom = input(f"Nouveau nom ({etudiant['nom']}) : ").strip().upper()
                if nouveau_nom:
                    etudiant['nom'] = nouveau_nom

                nouveau_prenom = input(f"Nouveau prÃ©nom ({etudiant['prenom']}) : ").strip().title()
                if nouveau_prenom:
                    etudiant['prenom'] = nouveau_prenom

                nouvel_age = input(f"Nouvel Ã¢ge ({etudiant['age']}) : ")
                if nouvel_age:
                    etudiant['age'] = int(nouvel_age)
                
                nouveau_sexe = input(f"Nouveau sexe ({etudiant['sexe']}) : ").strip().upper()
                if nouveau_sexe:
                    etudiant['sexe'] = nouveau_sexe
                
                nouvelle_nationalite = input(f"Nouvelle nationalitÃ© ({etudiant['nationalite']}) : ").strip().title()
                if nouvelle_nationalite:
                    etudiant['nationalite'] = nouvelle_nationalite
                
                # Modifier les notes
                print("\nðŸ“ Modification des notes (laissez vide pour ne pas modifier) :")
                nouvelles_notes = []
                for i in range(3):
                    nouvelle_note = input(f"Nouvelle note {i+1} ({etudiant['notes'][i]}) : ")
                    if nouvelle_note:
                        nouvelles_notes.append(float(nouvelle_note))
                    else:
                        nouvelles_notes.append(etudiant['notes'][i])
                
                etudiant['notes'] = nouvelles_notes
                
                # Recalculer tout
                calculer_moyennes(etudiants)
                determiner_mentions(etudiants)
                classer_par_merite(etudiants)
                
                print("âœ… Modification rÃ©ussie !")
            except ValueError:
                print("âŒ Erreur : Veuillez entrer des valeurs valides.")

        elif choix == '11':
            matricule = input("Entrez le matricule de l'Ã©tudiant Ã  supprimer : ")
            etudiant = rechercher_etudiant(etudiants, matricule)
            if not etudiant:
                print("âŒ Aucun Ã©tudiant trouvÃ© avec ce matricule.")
                continue
            
            confirmation = input(f"âš ï¸ ÃŠtes-vous sÃ»r de vouloir supprimer {etudiant['prenom']} {etudiant['nom']} ? (o/n) : ")
            if confirmation.lower() in ['o', 'oui', 'y', 'yes']:
                etudiants.remove(etudiant)
                classer_par_merite(etudiants)
                print(f"âœ… L'Ã©tudiant {etudiant['prenom']} {etudiant['nom']} a Ã©tÃ© supprimÃ©.")
            else:
                print("âŒ Suppression annulÃ©e.")

        elif choix == '12':
            exporter_resultats(etudiants)

        elif choix == '13':
            sauvegarder_donnees(etudiants)

        elif choix == '14':
            etudiants_restaures = restaurer_donnees()
            if etudiants_restaures:
                etudiants = etudiants_restaures

        elif choix == '15':
            afficher_aide()

        elif choix == '16':
            print("\nðŸ‘‹ Merci d'avoir utilisÃ© le systÃ¨me de gestion des Ã©tudiants !")
            print("ðŸ“Š Statistiques finales :")
            print(f"   â€¢ {len(etudiants)} Ã©tudiants gÃ©rÃ©s")
            print(f"   â€¢ Moyenne gÃ©nÃ©rale : {sum(e['moyenne'] for e in etudiants)/len(etudiants):.2f}/20")
            print("   â€¢ Au revoir ! ðŸ‘‹")
            break
        else:
            print("âŒ Choix invalide. Veuillez rÃ©essayer.")

if __name__ == "__main__":
    main() 
```
