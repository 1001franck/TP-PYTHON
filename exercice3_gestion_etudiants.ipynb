```python
import pandas as pd
from datetime import datetime
import json
import os

# =============================================================================
# EXERCICE 3 : GESTION DES √âTUDIANTS
# =============================================================================
# 
# Programme de gestion des √©tudiants avec calcul des moyennes et classement
# 
# Fonctionnalit√©s :
# - Gestion de 25 √©tudiants de 8 nationalit√©s diff√©rentes
# - Calcul automatique des moyennes et mentions
# - Classement par ordre de m√©rite avec attribution des rangs
# - Recherche avanc√©e (par matricule, nom, nationalit√©, etc.)
# - Statistiques de la classe et analyses de performance
# - Ajout, modification et suppression d'√©tudiants
# - Export des r√©sultats et sauvegarde des donn√©es
# 
# Syst√®me de mentions :
# < 10 : Insuffisant
# [10-12[ : Passable
# [12-14[ : Assez Bien
# [14-16[ : Bien
# [16-18[ : Tr√®s Bien
# [18-19.5[ : Excellent
# >= 19.5 : Honorable
# 
# Crit√®res d'√©valuation :
# - Admis : Moyenne >= 10/20
# - Redoublants : 8 <= Moyenne < 10/20
# - Exclus : Moyenne < 8/20
# 
# Menu principal (16 options) :
# 1-4 : Affichage et consultation
# 5-8 : Recherche et filtres
# 9-11 : Gestion des donn√©es
# 12-16 : Outils avanc√©s (export, sauvegarde, aide, quitter)
# 
# D√©velopp√© pour l'exercice 3 de gestion des √©tudiants par FOTSI, ILOUMBOU ET YOMBI
# =============================================================================

def get_donnees_initiales():
    """Retourne une liste d'√©tudiants pour commencer."""
    return [
        {
            'matricule': 'ETU001', 
            'nom': 'KOUASSI', 
            'prenom': 'Jean', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [15, 18, 12],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU002', 
            'nom': 'KOUAME', 
            'prenom': 'Marie', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [17, 19, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU003', 
            'nom': 'KOUADIO', 
            'prenom': 'Pierre', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [14, 16, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU004', 
            'nom': 'KOUASSI', 
            'prenom': 'Sophie', 
            'age': 20, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [16, 14, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU005', 
            'nom': 'KOUADIO', 
            'prenom': 'Paul', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Ivoirienne',
            'notes': [8, 9, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU006', 
            'nom': 'KOUAME', 
            'prenom': 'Alice', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'Ivoirienne',
            'notes': [19, 20, 19],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux √©tudiants - Cameroun
        {
            'matricule': 'ETU007', 
            'nom': 'MBALLA', 
            'prenom': 'David', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Camerounaise',
            'notes': [16, 17, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU008', 
            'nom': 'TCHOKOUANI', 
            'prenom': 'Fatou', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Camerounaise',
            'notes': [18, 19, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU009', 
            'nom': 'TAKAM', 
            'prenom': 'Franck', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Camerounaise',
            'notes': [12, 14, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux √©tudiants - Gabon
        {
            'matricule': 'ETU010', 
            'nom': 'ILOUMBOU', 
            'prenom': 'Junior', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [15, 16, 14],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU011', 
            'nom': 'NDONG', 
            'prenom': 'Thomas', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [9, 8, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU012', 
            'nom': 'YOMBI', 
            'prenom': 'Descartes', 
            'age': 24, 
            'sexe': 'M',
            'nationalite': 'Gabonaise',
            'notes': [20, 19, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux √©tudiants - Niger
        {
            'matricule': 'ETU013', 
            'nom': 'MAHAMADOU', 
            'prenom': 'Amina', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Nig√©rienne',
            'notes': [13, 15, 12],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU014', 
            'nom': 'ISSOUFOU', 
            'prenom': 'Moussa', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Nig√©rienne',
            'notes': [11, 9, 10],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU015', 
            'nom': 'TANDJA', 
            'prenom': 'Hassan', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'Nig√©rienne',
            'notes': [17, 18, 16],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # Nouveaux √©tudiants - Tchad
        {
            'matricule': 'ETU016', 
            'nom': 'DEBY', 
            'prenom': 'Zara', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'Tchadienne',
            'notes': [14, 13, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU017', 
            'nom': 'KABADI', 
            'prenom': 'Idriss', 
            'age': 22, 
            'sexe': 'M',
            'nationalite': 'Tchadienne',
            'notes': [6, 5, 7],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU018', 
            'nom': 'MAHAMAT', 
            'prenom': 'Fatima', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Tchadienne',
            'notes': [19, 20, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # 4 √©tudiants suppl√©mentaires
        {
            'matricule': 'ETU019', 
            'nom': 'KONATE', 
            'prenom': 'Moussa', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'Malienne',
            'notes': [14, 16, 15],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU020', 
            'nom': 'DIALLO', 
            'prenom': 'Aissatou', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'Guin√©enne',
            'notes': [8, 7, 6],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU021', 
            'nom': 'TOURE', 
            'prenom': 'S√©kou', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'Malienne',
            'notes': [18, 19, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU022', 
            'nom': 'CAMARA', 
            'prenom': 'Mariama', 
            'age': 18, 
            'sexe': 'F',
            'nationalite': 'Guin√©enne',
            'notes': [20, 19, 20],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        # 3 √©tudiants suppl√©mentaires
        {
            'matricule': 'ETU023', 
            'nom': 'SALL', 
            'prenom': 'Macky', 
            'age': 20, 
            'sexe': 'M',
            'nationalite': 'S√©n√©galaise',
            'notes': [16, 15, 17],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU024', 
            'nom': 'WADE', 
            'prenom': 'Aminata', 
            'age': 19, 
            'sexe': 'F',
            'nationalite': 'S√©n√©galaise',
            'notes': [12, 11, 13],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        },
        {
            'matricule': 'ETU025', 
            'nom': 'DIOP', 
            'prenom': 'Cheikh', 
            'age': 21, 
            'sexe': 'M',
            'nationalite': 'S√©n√©galaise',
            'notes': [19, 20, 18],
            'moyenne': 0,
            'mention': '',
            'rang': 0
        }
    ]

def calculer_moyennes(liste_etudiants):
    """Calcule la moyenne de chaque √©tudiant."""
    for etudiant in liste_etudiants:
        if len(etudiant['notes']) == 3:
            etudiant['moyenne'] = sum(etudiant['notes']) / 3
        else:
            etudiant['moyenne'] = 0
    return liste_etudiants

def determiner_mentions(liste_etudiants):
    """D√©termine la mention de chaque √©tudiant."""
    for etudiant in liste_etudiants:
        moyenne = etudiant['moyenne']
        if moyenne < 10:
            etudiant['mention'] = 'Insuffisant'
        elif moyenne < 12:
            etudiant['mention'] = 'Passable'
        elif moyenne < 14:
            etudiant['mention'] = 'Assez Bien'
        elif moyenne < 16:
            etudiant['mention'] = 'Bien'
        elif moyenne < 18:
            etudiant['mention'] = 'Tr√®s Bien'
        elif moyenne < 19.5:
            etudiant['mention'] = 'Excellent'
        else:
            etudiant['mention'] = 'Honorable'
    return liste_etudiants

def classer_par_merite(liste_etudiants):
    """Classe les √©tudiants par ordre de m√©rite et affecte le rang."""
    # Trier par moyenne d√©croissante
    liste_triee = sorted(liste_etudiants, key=lambda x: x['moyenne'], reverse=True)
    
    # Affecter le rang
    for i, etudiant in enumerate(liste_triee, 1):
        etudiant['rang'] = i
    
    return liste_triee

def afficher_informations_etudiants(liste_etudiants):
    """Affiche les informations de base des √©tudiants."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # En-t√™tes des colonnes
    headers = ['Matricule', 'NOM', 'Pr√©nom', 'Age', 'Sexe', 'Nationalit√©', 'Notes']
    
    # Pr√©parer les donn√©es pour l'affichage
    data = []
    for etudiant in liste_etudiants:
        notes_str = f"{etudiant['notes'][0]}, {etudiant['notes'][1]}, {etudiant['notes'][2]}"
        
        data.append([
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            str(etudiant['age']),
            etudiant['sexe'],
            etudiant['nationalite'],
            notes_str
        ])
    
    # Calculer la largeur de chaque colonne
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour cr√©er une ligne de s√©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-t√™tes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # Donn√©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def afficher_resultats_etudiants(liste_etudiants):
    """Affiche les r√©sultats des √©tudiants (moyennes, mentions, classement)."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # Trier les √©tudiants par rang (ordre de m√©rite)
    etudiants_tries = sorted(liste_etudiants, key=lambda x: x['rang'])
    
    # En-t√™tes des colonnes
    headers = ['Rang', 'Matricule', 'NOM', 'Pr√©nom', 'Moyenne', 'Mention', 'Statut']
    
    # Pr√©parer les donn√©es pour l'affichage
    data = []
    for etudiant in etudiants_tries:
        # D√©terminer le statut
        moyenne = etudiant['moyenne']
        if moyenne >= 10:
            statut = "Admis"
        elif moyenne >= 8:
            statut = "Redouble"
        else:
            statut = "Exclu"
        
        data.append([
            str(etudiant['rang']),
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            f"{etudiant['moyenne']:.2f}",
            f"{etudiant['mention']}",
            statut
        ])
    
    # Calculer la largeur de chaque colonne avec une largeur minimale pour Statut
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        
        # Largeur minimale pour la colonne Statut
        if header == 'Statut':
            max_width = max(max_width, 12)  # Au moins 12 caract√®res pour "Redouble"
        
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour cr√©er une ligne de s√©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-t√™tes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # Donn√©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def afficher_admis(liste_etudiants):
    """Affiche les √©tudiants admis (moyenne >= 10)."""
    admis = [etudiant for etudiant in liste_etudiants if etudiant['moyenne'] >= 10]
    if admis:
        print(f"\n=== √âTUDIANTS ADMIS ({len(admis)} √©tudiants) ===")
        afficher_informations_etudiants(admis)
    else:
        print("\nAucun √©tudiant admis.")

def afficher_redoublants(liste_etudiants):
    """Affiche les redoublants (moyenne entre 8 et 10)."""
    redoublants = [etudiant for etudiant in liste_etudiants if 8 <= etudiant['moyenne'] < 10]
    if redoublants:
        print(f"\n=== REDOUBLANTS ({len(redoublants)} √©tudiants) ===")
        afficher_informations_etudiants(redoublants)
    else:
        print("\nAucun redoublant.")

def afficher_exclus(liste_etudiants):
    """Affiche les exclus (moyenne < 8)."""
    exclus = [etudiant for etudiant in liste_etudiants if etudiant['moyenne'] < 8]
    if exclus:
        print(f"\n=== EXCLUS ({len(exclus)} √©tudiants) ===")
        afficher_informations_etudiants(exclus)
    else:
        print("\nAucun √©tudiant exclu.")

def afficher_etudiants_alphab√©tique(liste_etudiants):
    """Affiche les √©tudiants par ordre alphab√©tique (nom, puis pr√©nom)."""
    if not liste_etudiants:
        print("\nLa liste est vide.")
        return
    
    # Trier les √©tudiants par ordre alphab√©tique (nom, puis pr√©nom)
    etudiants_tries = sorted(liste_etudiants, key=lambda x: (x['nom'], x['prenom']))
    
    # En-t√™tes des colonnes
    headers = ['N¬∞', 'Matricule', 'NOM', 'Pr√©nom', 'Age', 'Sexe', 'Nationalit√©', 'Notes']
    
    # Pr√©parer les donn√©es pour l'affichage
    data = []
    for i, etudiant in enumerate(etudiants_tries, 1):
        notes_str = f"{etudiant['notes'][0]}, {etudiant['notes'][1]}, {etudiant['notes'][2]}"
        
        data.append([
            str(i),
            etudiant['matricule'],
            etudiant['nom'],
            etudiant['prenom'],
            str(etudiant['age']),
            etudiant['sexe'],
            etudiant['nationalite'],
            notes_str
        ])
    
    # Calculer la largeur de chaque colonne
    col_widths = []
    for i, header in enumerate(headers):
        max_width = len(header)
        for row in data:
            max_width = max(max_width, len(row[i]))
        col_widths.append(max_width + 2)  # +2 pour l'espacement
    
    # Fonction pour centrer le texte
    def center_text(text, width):
        return text.center(width)
    
    # Fonction pour cr√©er une ligne de s√©paration
    def separator():
        return '+' + '+'.join('-' * width for width in col_widths) + '+'
    
    # Afficher le tableau
    print("\n" + separator())
    
    # En-t√™tes
    header_line = '|'
    for i, header in enumerate(headers):
        header_line += center_text(header, col_widths[i]) + '|'
    print(header_line)
    print(separator())
    
    # Donn√©es
    for row in data:
        data_line = '|'
        for i, cell in enumerate(row):
            data_line += center_text(cell, col_widths[i]) + '|'
        print(data_line)
    
    print(separator())

def calculer_statistiques(liste_etudiants):
    """Calcule les statistiques avanc√©es de la classe."""
    if not liste_etudiants:
        print("\nAucun √©tudiant pour calculer les statistiques.")
        return
    
    # Statistiques de base
    nb_total = len(liste_etudiants)
    moyennes = [etudiant['moyenne'] for etudiant in liste_etudiants]
    moyenne_generale = sum(moyennes) / nb_total
    note_min = min(moyennes)
    note_max = max(moyennes)
    
    # √âcart-type
    variance = sum((m - moyenne_generale) ** 2 for m in moyennes) / nb_total
    ecart_type = variance ** 0.5
    
    # R√©partition par nationalit√©
    nationalites = {}
    for etudiant in liste_etudiants:
        nat = etudiant['nationalite']
        nationalites[nat] = nationalites.get(nat, 0) + 1
    
    # R√©partition par sexe
    hommes = len([e for e in liste_etudiants if e['sexe'] == 'M'])
    femmes = len([e for e in liste_etudiants if e['sexe'] == 'F'])
    
    # R√©partition par mention
    mentions = {}
    for etudiant in liste_etudiants:
        mention = etudiant['mention']
        mentions[mention] = mentions.get(mention, 0) + 1
    
    # R√©partition par statut
    admis = len([e for e in liste_etudiants if e['moyenne'] >= 10])
    redoublants = len([e for e in liste_etudiants if 8 <= e['moyenne'] < 10])
    exclus = len([e for e in liste_etudiants if e['moyenne'] < 8])
    
    # Affichage des statistiques
    print("\n" + "="*60)
    print("STATISTIQUES DE LA CLASSE")
    print("="*60)
    
    print(f"\nEffectif total : {nb_total} √©tudiants")
    print(f"Note la plus basse : {note_min:.2f}/20")
    print(f"Note la plus haute : {note_max:.2f}/20")
    print(f"\nR√©partition par sexe :")
    print(f"   Hommes : {hommes} ({hommes/nb_total*100:.1f}%)")
    print(f"   Femmes : {femmes} ({femmes/nb_total*100:.1f}%)")
    
    print(f"\nR√©partition par nationalit√© :")
    for nat, nb in sorted(nationalites.items()):
        print(f"   {nat} : {nb} √©tudiant(s) ({nb/nb_total*100:.1f}%)")
    
    print(f"\nR√©partition par mention :")
    for mention, nb in sorted(mentions.items()):
        print(f"   {mention} : {nb} √©tudiant(s) ({nb/nb_total*100:.1f}%)")
    
    print(f"\nR√©partition par statut :")
    print(f"   Admis : {admis} √©tudiant(s) ({admis/nb_total*100:.1f}%)")
    print(f"   Redoublants : {redoublants} √©tudiant(s) ({redoublants/nb_total*100:.1f}%)")
    print(f"   Exclus : {exclus} √©tudiant(s) ({exclus/nb_total*100:.1f}%)")
    
    print("="*60)

def rechercher_etudiant_complet(liste_etudiants):
    """Recherche compl√®te d'√©tudiants avec plusieurs options."""
    print("\n--- RECHERCHE D'√âTUDIANTS ---")
    print("1. Rechercher par matricule")
    print("2. Rechercher par nom/pr√©nom")
    print("3. Rechercher par nationalit√©")
    print("4. Rechercher par mention")
    print("5. Rechercher par statut")
    print("6. Rechercher par tranche d'√¢ge")
    print("7. Retour au menu principal")
    
    choix_recherche = input("Votre choix : ")
    
    if choix_recherche == '1':
        matricule = input("Entrez le matricule de l'√©tudiant √† rechercher : ")
        etudiant_trouve = rechercher_etudiant(liste_etudiants, matricule)
        if etudiant_trouve:
            print(f"\n=== D√âTAILS DE L'√âTUDIANT ===")
            print(f"Matricule : {etudiant_trouve['matricule']}")
            print(f"Nom : {etudiant_trouve['nom']}")
            print(f"Pr√©nom : {etudiant_trouve['prenom']}")
            print(f"Age : {etudiant_trouve['age']} ans")
            print(f"Sexe : {etudiant_trouve['sexe']}")
            print(f"Nationalit√© : {etudiant_trouve['nationalite']}")
            print(f"Notes : {etudiant_trouve['notes'][0]}, {etudiant_trouve['notes'][1]}, {etudiant_trouve['notes'][2]}")
            print(f"Moyenne : {etudiant_trouve['moyenne']:.2f}")
            print(f"Mention : {etudiant_trouve['mention']}")
            print(f"Rang : {etudiant_trouve['rang']}")
        else:
            print("Aucun √©tudiant trouv√© avec ce matricule.")
    
    elif choix_recherche == '2':
        terme = input("Entrez le nom ou pr√©nom √† rechercher : ").upper()
        resultats = []
        for etudiant in liste_etudiants:
            if terme in etudiant['nom'].upper() or terme in etudiant['prenom'].upper():
                resultats.append(etudiant)
        
        if resultats:
            print(f"\n{len(resultats)} √©tudiant(s) trouv√©(s) :")
            afficher_informations_etudiants(resultats)
        else:
            print("Aucun √©tudiant trouv√©.")
    
    elif choix_recherche == '3':
        print("\nNationalit√©s disponibles :")
        nationalites = list(set(etudiant['nationalite'] for etudiant in liste_etudiants))
        for i, nat in enumerate(nationalites, 1):
            print(f"{i}. {nat}")
        
        try:
            choix = int(input("Choisissez le num√©ro de la nationalit√© : ")) - 1
            if 0 <= choix < len(nationalites):
                nationalite = nationalites[choix]
                resultats = [e for e in liste_etudiants if e['nationalite'] == nationalite]
                print(f"\n{len(resultats)} √©tudiant(s) {nationalite}(s) :")
                afficher_informations_etudiants(resultats)
            else:
                print("Choix invalide.")
        except ValueError:
            print("Veuillez entrer un num√©ro valide.")
    
    elif choix_recherche == '4':
        print("\nMentions disponibles :")
        mentions = ['Insuffisant', 'Passable', 'Assez Bien', 'Bien', 'Tr√®s Bien', 'Excellent', 'Honorable']
        for i, mention in enumerate(mentions, 1):
            print(f"{i}. {mention}")
        
        try:
            choix = int(input("Choisissez le num√©ro de la mention : ")) - 1
            if 0 <= choix < len(mentions):
                mention = mentions[choix]
                resultats = [e for e in liste_etudiants if e['mention'] == mention]
                if resultats:
                    print(f"\n{len(resultats)} √©tudiant(s) avec mention '{mention}' :")
                    afficher_informations_etudiants(resultats)
                else:
                    print(f"Aucun √©tudiant avec la mention '{mention}'.")
            else:
                print("Choix invalide.")
        except ValueError:
            print("Veuillez entrer un num√©ro valide.")
    
    elif choix_recherche == '5':
        print("\nStatuts disponibles :")
        print("1. Admis (moyenne >= 10)")
        print("2. Redoublants (8 <= moyenne < 10)")
        print("3. Exclus (moyenne < 8)")
        
        choix = input("Choisissez le statut (1-3) : ")
        if choix == '1':
            resultats = [e for e in liste_etudiants if e['moyenne'] >= 10]
            print(f"\n{len(resultats)} √©tudiant(s) admis :")
        elif choix == '2':
            resultats = [e for e in liste_etudiants if 8 <= e['moyenne'] < 10]
            print(f"\n{len(resultats)} redoublant(s) :")
        elif choix == '3':
            resultats = [e for e in liste_etudiants if e['moyenne'] < 8]
            print(f"\n{len(resultats)} √©tudiant(s) exclu(s) :")
        else:
            print("Choix invalide.")
            return
        
        if resultats:
            afficher_informations_etudiants(resultats)
        else:
            print("Aucun √©tudiant trouv√©.")
    
    elif choix_recherche == '6':
        try:
            age_min = int(input("√Çge minimum : "))
            age_max = int(input("√Çge maximum : "))
            resultats = [e for e in liste_etudiants if age_min <= e['age'] <= age_max]
            
            if resultats:
                print(f"\n{len(resultats)} √©tudiant(s) entre {age_min} et {age_max} ans :")
                afficher_informations_etudiants(resultats)
            else:
                print("Aucun √©tudiant trouv√© dans cette tranche d'√¢ge.")
        except ValueError:
            print("Veuillez entrer des √¢ges valides.")
    
    elif choix_recherche == '7':
        return
    
    else:
        print("Choix invalide.")

def valider_donnees_etudiant(matricule, nom, prenom, age, sexe, nationalite, notes):
    """Valide les donn√©es d'un √©tudiant avant ajout/modification."""
    erreurs = []
    
    # Validation du matricule
    if not matricule or len(matricule.strip()) < 3:
        erreurs.append("Le matricule doit contenir au moins 3 caract√®res")
    
    # Validation du nom et pr√©nom
    if not nom or len(nom.strip()) < 2:
        erreurs.append("Le nom doit contenir au moins 2 caract√®res")
    if not prenom or len(prenom.strip()) < 2:
        erreurs.append("Le pr√©nom doit contenir au moins 2 caract√®res")
    
    # Validation de l'√¢ge
    if not isinstance(age, int) or age < 15 or age > 35:
        erreurs.append("L'√¢ge doit √™tre entre 15 et 35 ans")
    
    # Validation du sexe
    if sexe not in ['M', 'F']:
        erreurs.append("Le sexe doit √™tre 'M' ou 'F'")
    
    # Validation de la nationalit√©
    if not nationalite or len(nationalite.strip()) < 3:
        erreurs.append("La nationalit√© doit contenir au moins 3 caract√®res")
    
    # Validation des notes
    if len(notes) != 3:
        erreurs.append("Il faut exactement 3 notes")
    else:
        for i, note in enumerate(notes, 1):
            if not isinstance(note, (int, float)) or note < 0 or note > 20:
                erreurs.append(f"La note {i} doit √™tre entre 0 et 20")
    
    return erreurs

def exporter_resultats(liste_etudiants):
    """Exporte les r√©sultats dans un fichier texte."""
    try:
        nom_fichier = f"resultats_etudiants_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
        
        with open(nom_fichier, 'w', encoding='utf-8') as f:
            f.write("="*80 + "\n")
            f.write("RAPPORT DES R√âSULTATS - GESTION DES √âTUDIANTS\n")
            f.write("="*80 + "\n\n")
            
            # Informations g√©n√©rales
            f.write(f"Date de g√©n√©ration : {datetime.now().strftime('%d/%m/%Y √† %H:%M')}\n")
            f.write(f"Nombre total d'√©tudiants : {len(liste_etudiants)}\n\n")
            
            # Statistiques
            moyennes = [e['moyenne'] for e in liste_etudiants]
            moyenne_generale = sum(moyennes) / len(moyennes)
            f.write(f"Moyenne g√©n√©rale de la classe : {moyenne_generale:.2f}/20\n\n")
            
            # Classement complet
            f.write("CLASSEMENT PAR ORDRE DE M√âRITE\n")
            f.write("-"*80 + "\n")
            f.write(f"{'Rang':<5} {'Matricule':<10} {'Nom':<15} {'Pr√©nom':<15} {'Moyenne':<8} {'Mention':<12} {'Statut':<10}\n")
            f.write("-"*80 + "\n")
            
            etudiants_tries = sorted(liste_etudiants, key=lambda x: x['rang'])
            for etudiant in etudiants_tries:
                statut = "Admis" if etudiant['moyenne'] >= 10 else "Redouble" if etudiant['moyenne'] >= 8 else "Exclu"
                f.write(f"{etudiant['rang']:<5} {etudiant['matricule']:<10} {etudiant['nom']:<15} {etudiant['prenom']:<15} {etudiant['moyenne']:<8.2f} {etudiant['mention']:<12} {statut:<10}\n")
            
            f.write("\n" + "="*80 + "\n")
            f.write("FIN DU RAPPORT\n")
            f.write("="*80 + "\n")
        
        print(f"Rapport export√© avec succ√®s : {nom_fichier}")
        return True
        
    except Exception as e:
        print(f"Erreur lors de l'export : {e}")
        return False

def sauvegarder_donnees(liste_etudiants):
    """Sauvegarde les donn√©es dans un fichier JSON."""
    try:
        nom_fichier = f"backup_etudiants_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
        
        # Pr√©parer les donn√©es pour la sauvegarde
        donnees_sauvegarde = []
        for etudiant in liste_etudiants:
            donnees_sauvegarde.append({
                'matricule': etudiant['matricule'],
                'nom': etudiant['nom'],
                'prenom': etudiant['prenom'],
                'age': etudiant['age'],
                'sexe': etudiant['sexe'],
                'nationalite': etudiant['nationalite'],
                'notes': etudiant['notes']
            })
        
        with open(nom_fichier, 'w', encoding='utf-8') as f:
            json.dump(donnees_sauvegarde, f, indent=2, ensure_ascii=False)
        
        print(f"Sauvegarde cr√©√©e avec succ√®s : {nom_fichier}")
        return True
        
    except Exception as e:
        print(f"Erreur lors de la sauvegarde : {e}")
        return False

def restaurer_donnees():
    """Restaure les donn√©es depuis un fichier JSON."""
    try:
        # Lister les fichiers de sauvegarde disponibles
        fichiers_backup = [f for f in os.listdir('.') if f.startswith('backup_etudiants_') and f.endswith('.json')]
        
        if not fichiers_backup:
            print("Aucun fichier de sauvegarde trouv√©.")
            return None
        
        print("\nüìÅ Fichiers de sauvegarde disponibles :")
        for i, fichier in enumerate(sorted(fichiers_backup, reverse=True), 1):
            print(f"   {i}. {fichier}")
        
        choix = input("\nChoisissez le num√©ro du fichier √† restaurer : ")
        try:
            index = int(choix) - 1
            if 0 <= index < len(fichiers_backup):
                fichier_choisi = sorted(fichiers_backup, reverse=True)[index]
                
                with open(fichier_choisi, 'r', encoding='utf-8') as f:
                    donnees = json.load(f)
                
                # Convertir les donn√©es en format √©tudiant
                etudiants_restaures = []
                for donnee in donnees:
                    etudiant = {
                        'matricule': donnee['matricule'],
                        'nom': donnee['nom'],
                        'prenom': donnee['prenom'],
                        'age': donnee['age'],
                        'sexe': donnee['sexe'],
                        'nationalite': donnee['nationalite'],
                        'notes': donnee['notes'],
                        'moyenne': 0,
                        'mention': '',
                        'rang': 0
                    }
                    etudiants_restaures.append(etudiant)
                
                # Recalculer tout
                calculer_moyennes(etudiants_restaures)
                determiner_mentions(etudiants_restaures)
                classer_par_merite(etudiants_restaures)
                
                print(f"Donn√©es restaur√©es depuis {fichier_choisi}")
                print(f"{len(etudiants_restaures)} √©tudiants charg√©s")
                return etudiants_restaures
            else:
                print("Choix invalide.")
                return None
        except ValueError:
            print("Veuillez entrer un num√©ro valide.")
            return None
            
    except Exception as e:
        print(f"Erreur lors de la restauration : {e}")
        return None

def analyser_performances(liste_etudiants):
    """Analyse d√©taill√©e des performances de la classe."""
    if not liste_etudiants:
        print("Aucun √©tudiant pour analyser.")
        return
    
    print("\n" + "="*60)
    print("ANALYSE D√âTAILL√âE DES PERFORMANCES")
    print("="*60)
    
    # Statistiques de base
    nb_total = len(liste_etudiants)
    moyennes = [e['moyenne'] for e in liste_etudiants]
    
    # Meilleurs et moins bons
    meilleur = max(liste_etudiants, key=lambda x: x['moyenne'])
    moins_bon = min(liste_etudiants, key=lambda x: x['moyenne'])
    
    # R√©partition par tranches de notes
    excellent = len([e for e in liste_etudiants if e['moyenne'] >= 16])
    tres_bien = len([e for e in liste_etudiants if 14 <= e['moyenne'] < 16])
    bien = len([e for e in liste_etudiants if 12 <= e['moyenne'] < 14])
    passable = len([e for e in liste_etudiants if 10 <= e['moyenne'] < 12])
    insuffisant = len([e for e in liste_etudiants if e['moyenne'] < 10])
    
    print(f"\nMEILLEURS √âL√àVES :")
    print(f"   ‚Ä¢ 1er : {meilleur['prenom']} {meilleur['nom']} - {meilleur['moyenne']:.2f}/20 ({meilleur['mention']})")
    
    # Trouver le 2√®me et 3√®me
    etudiants_tries = sorted(liste_etudiants, key=lambda x: x['moyenne'], reverse=True)
    if len(etudiants_tries) >= 2:
        deuxieme = etudiants_tries[1]
        print(f"   ‚Ä¢ 2√®me : {deuxieme['prenom']} {deuxieme['nom']} - {deuxieme['moyenne']:.2f}/20 ({deuxieme['mention']})")
    if len(etudiants_tries) >= 3:
        troisieme = etudiants_tries[2]
        print(f"   ‚Ä¢ 3√®me : {troisieme['prenom']} {troisieme['nom']} - {troisieme['moyenne']:.2f}/20 ({troisieme['mention']})")
    
    print(f"\n√âL√àVES EN DIFFICULT√â :")
    print(f"   ‚Ä¢ Plus faible : {moins_bon['prenom']} {moins_bon['nom']} - {moins_bon['moyenne']:.2f}/20 ({moins_bon['mention']})")
    
    print(f"\nR√âPARTITION PAR NIVEAU :")
    print(f"   ‚Ä¢ Excellent (‚â•16) : {excellent} √©tudiant(s) ({excellent/nb_total*100:.1f}%)")
    print(f"   ‚Ä¢ Tr√®s Bien (14-16) : {tres_bien} √©tudiant(s) ({tres_bien/nb_total*100:.1f}%)")
    print(f"   ‚Ä¢ Bien (12-14) : {bien} √©tudiant(s) ({bien/nb_total*100:.1f}%)")
    print(f"   ‚Ä¢ Passable (10-12) : {passable} √©tudiant(s) ({passable/nb_total*100:.1f}%)")
    print(f"   ‚Ä¢ Insuffisant (<10) : {insuffisant} √©tudiant(s) ({insuffisant/nb_total*100:.1f}%)")
    
    # Analyse par nationalit√©
    print(f"\nR√âPARTITION PAR NATIONALIT√â :")
    nationalites = {}
    for etudiant in liste_etudiants:
        nat = etudiant['nationalite']
        if nat not in nationalites:
            nationalites[nat] = {'total': 0, 'moyennes': []}
        nationalites[nat]['total'] += 1
        nationalites[nat]['moyennes'].append(etudiant['moyenne'])
    
    for nat, data in sorted(nationalites.items()):
        moyenne_nat = sum(data['moyennes']) / len(data['moyennes'])
        print(f"   ‚Ä¢ {nat} : {data['total']} √©tudiant(s) - Moyenne : {moyenne_nat:.2f}/20")
    
    print("="*60)

def afficher_aide():
    """Affiche l'aide et les instructions d'utilisation."""
    print("\n" + "="*60)
    print(" AIDE ET INSTRUCTIONS")
    print("="*60)
    
    print("\nFONCTIONNALIT√âS PRINCIPALES :")
    print("   ‚Ä¢ Gestion compl√®te des √©tudiants (25 √©tudiants de 8 nationalit√©s)")
    print("   ‚Ä¢ Calcul automatique des moyennes et mentions")
    print("   ‚Ä¢ Classement par ordre de m√©rite")
    print("   ‚Ä¢ Recherche avanc√©e multi-crit√®res")
    print("   ‚Ä¢ Statistiques d√©taill√©es de la classe")
    
    print("\n SYST√àME DE MENTIONS :")
    print("   ‚Ä¢ < 10 : Insuffisant")
    print("   ‚Ä¢ [10-12[ : Passable")
    print("   ‚Ä¢ [12-14[ : Assez Bien")
    print("   ‚Ä¢ [14-16[ : Bien")
    print("   ‚Ä¢ [16-18[ : Tr√®s Bien")
    print("   ‚Ä¢ [18-19.5[ : Excellent")
    print("   ‚Ä¢ ‚â• 19.5 : Honorable")
    
    print("\nCRIT√àRES D'√âVALUATION :")
    print("   ‚Ä¢ Admis : Moyenne ‚â• 10/20")
    print("   ‚Ä¢ Redoublants : 8 ‚â§ Moyenne < 10/20")
    print("   ‚Ä¢ Exclus : Moyenne < 8/20")
    
    print("\nCONSEILS D'UTILISATION :")
    print("   ‚Ä¢ Utilisez les options 1-3 pour consulter les donn√©es")
    print("   ‚Ä¢ L'option 4 permet une recherche compl√®te")
    print("   ‚Ä¢ Sauvegardez r√©guli√®rement vos donn√©es (option 12)")
    print("   ‚Ä¢ Exportez les r√©sultats pour les partager (option 11)")
    
    print("\nüîß VALIDATION DES DONN√âES :")
    print("   ‚Ä¢ Matricule : minimum 3 caract√®res, unique")
    print("   ‚Ä¢ Nom/Pr√©nom : minimum 2 caract√®res")
    print("   ‚Ä¢ √Çge : entre 15 et 35 ans")
    print("   ‚Ä¢ Notes : entre 0 et 20 sur 20")
    
    print("="*60)

def afficher_menu_principal():
    """Affiche le menu principal avec une meilleure organisation."""
    print("\n" + "="*60)
    print("SYSTEME DE GESTION DES ETUDIANTS")
    print("="*60)
    print("\nAFFICHAGE ET CONSULTATION")
    print("  1. Liste des √©tudiants (ordre alphab√©tique)")
    print("  2. Classement par ordre de m√©rite")
    print("  3. Statistiques de la classe")
    print("  4. Analyse d√©taill√©e des performances")
    print("\nRECHERCHE ET FILTRES")
    print("  5. Rechercher un √©tudiant")
    print("  6. Afficher les admis")
    print("  7. Afficher les redoublants")
    print("  8. Afficher les exclus")
    print("\n GESTION DES DONN√âES")
    print("  9. Ajouter un √©tudiant")
    print("  10. Modifier un √©tudiant")
    print("  11. Supprimer un √©tudiant")
    print("\nOUTILS AVANCES")
    print("  12. Exporter les r√©sultats")
    print("  13. Sauvegarder les donn√©es")
    print("  14. Restaurer les donn√©es")
    print("  15. Aide et instructions")
    print("  16. Quitter")
    print("\n" + "="*60)

def menu():
    """Affiche le menu principal et retourne le choix de l'utilisateur."""
    afficher_menu_principal()
    return input("Votre choix : ")

def rechercher_etudiant(liste_etudiants, matricule):
    """Recherche un √©tudiant par son matricule."""
    for etudiant in liste_etudiants:
        if etudiant['matricule'].lower() == matricule.lower():
            return etudiant
    return None

def ajouter_etudiant(liste_etudiants):
    """Ajoute un nouvel √©tudiant √† la liste avec validation."""
    print("\n" + "="*50)
    print("‚ûï AJOUT D'UN NOUVEL √âTUDIANT")
    print("="*50)
    
    try:
        # Saisie des donn√©es
        matricule = input("Matricule : ").strip()
        nom = input("Nom : ").strip().upper()
        prenom = input("Pr√©nom : ").strip().title()
        age = int(input("Age : "))
        sexe = input("Sexe (M/F) : ").strip().upper()
        nationalite = input("Nationalit√© : ").strip().title()
        
        print("\n Saisie des 3 notes (sur 20) :")
        notes = []
        for i in range(3):
            while True:
                try:
                    note = float(input(f"Note {i+1} : "))
                    if 0 <= note <= 20:
                        notes.append(note)
                        break
                    else:
                        print("La note doit √™tre entre 0 et 20.")
                except ValueError:
                    print("Veuillez entrer un nombre valide.")

        # Validation des donn√©es
        erreurs = valider_donnees_etudiant(matricule, nom, prenom, age, sexe, nationalite, notes)
        
        if erreurs:
            print("\nErreurs de validation :")
            for erreur in erreurs:
                print(f"   ‚Ä¢ {erreur}")
            return

        # V√©rifier si le matricule existe d√©j√†
        if rechercher_etudiant(liste_etudiants, matricule):
            print(f"Le matricule {matricule} existe d√©j√†.")
            return

        # Cr√©er le nouvel √©tudiant
        nouvel_etudiant = {
            'matricule': matricule,
            'nom': nom,
            'prenom': prenom,
            'age': age,
            'sexe': sexe,
            'nationalite': nationalite,
            'notes': notes,
            'moyenne': 0,
            'mention': '',
            'rang': 0
        }
        
        liste_etudiants.append(nouvel_etudiant)
        
        # Recalculer tout
        calculer_moyennes(liste_etudiants)
        determiner_mentions(liste_etudiants)
        classer_par_merite(liste_etudiants)
        
        print(f"\nL'√©tudiant {prenom} {nom} a √©t√© ajout√© avec succ√®s !")
        print(f"Moyenne : {nouvel_etudiant['moyenne']:.2f}/20")
        print(f"Mention : {nouvel_etudiant['mention']}")
        print(f"Rang : {nouvel_etudiant['rang']}")

    except ValueError:
        print("Erreur : L'√¢ge doit √™tre un nombre entier.")
    except Exception as e:
        print(f"Une erreur inattendue est survenue : {e}")

def main():
    etudiants = get_donnees_initiales()
    
    # Initialisation : calculer moyennes, mentions et classement
    calculer_moyennes(etudiants)
    determiner_mentions(etudiants)
    classer_par_merite(etudiants)

    print("Bienvenue dans le syst√®me de gestion des √©tudiants !")
    print(f"{len(etudiants)} √©tudiants charg√©s avec succ√®s.")

    while True:
        choix = menu()
        
        if choix == '1':
            afficher_etudiants_alphab√©tique(etudiants)
        
        elif choix == '2':
            afficher_resultats_etudiants(etudiants)
        
        elif choix == '3':
            calculer_statistiques(etudiants)
        
        elif choix == '4':
            analyser_performances(etudiants)
        
        elif choix == '5':
            rechercher_etudiant_complet(etudiants)
        
        elif choix == '6':
            afficher_admis(etudiants)
        
        elif choix == '7':
            afficher_redoublants(etudiants)
        
        elif choix == '8':
            afficher_exclus(etudiants)
        
        elif choix == '9':
            ajouter_etudiant(etudiants)
        
        elif choix == '10':
            matricule = input("Entrez le matricule de l'√©tudiant √† modifier : ")
            etudiant = rechercher_etudiant(etudiants, matricule)
            if not etudiant:
                print("Aucun √©tudiant trouv√© avec ce matricule.")
                continue
            
            print(f"\n Modification de {etudiant['prenom']} {etudiant['nom']}")
            print("Laissez vide pour ne pas modifier.")
            
            try:
                nouveau_nom = input(f"Nouveau nom ({etudiant['nom']}) : ").strip().upper()
                if nouveau_nom:
                    etudiant['nom'] = nouveau_nom

                nouveau_prenom = input(f"Nouveau pr√©nom ({etudiant['prenom']}) : ").strip().title()
                if nouveau_prenom:
                    etudiant['prenom'] = nouveau_prenom

                nouvel_age = input(f"Nouvel √¢ge ({etudiant['age']}) : ")
                if nouvel_age:
                    etudiant['age'] = int(nouvel_age)
                
                nouveau_sexe = input(f"Nouveau sexe ({etudiant['sexe']}) : ").strip().upper()
                if nouveau_sexe:
                    etudiant['sexe'] = nouveau_sexe
                
                nouvelle_nationalite = input(f"Nouvelle nationalit√© ({etudiant['nationalite']}) : ").strip().title()
                if nouvelle_nationalite:
                    etudiant['nationalite'] = nouvelle_nationalite
                
                # Modifier les notes
                print("\n Modification des notes (laissez vide pour ne pas modifier) :")
                nouvelles_notes = []
                for i in range(3):
                    nouvelle_note = input(f"Nouvelle note {i+1} ({etudiant['notes'][i]}) : ")
                    if nouvelle_note:
                        nouvelles_notes.append(float(nouvelle_note))
                    else:
                        nouvelles_notes.append(etudiant['notes'][i])
                
                etudiant['notes'] = nouvelles_notes
                
                # Recalculer tout
                calculer_moyennes(etudiants)
                determiner_mentions(etudiants)
                classer_par_merite(etudiants)
                
                print("Modification r√©ussie !")
            except ValueError:
                print("Erreur : Veuillez entrer des valeurs valides.")

        elif choix == '11':
            matricule = input("Entrez le matricule de l'√©tudiant √† supprimer : ")
            etudiant = rechercher_etudiant(etudiants, matricule)
            if not etudiant:
                print("Aucun √©tudiant trouv√© avec ce matricule.")
                continue
            
            confirmation = input(f"√ätes-vous s√ªr de vouloir supprimer {etudiant['prenom']} {etudiant['nom']} ? (o/n) : ")
            if confirmation.lower() in ['o', 'oui', 'y', 'yes']:
                etudiants.remove(etudiant)
                classer_par_merite(etudiants)
                print(f"L'√©tudiant {etudiant['prenom']} {etudiant['nom']} a √©t√© supprim√©.")
            else:
                print("Suppression annul√©e.")

        elif choix == '12':
            exporter_resultats(etudiants)

        elif choix == '13':
            sauvegarder_donnees(etudiants)

        elif choix == '14':
            etudiants_restaures = restaurer_donnees()
            if etudiants_restaures:
                etudiants = etudiants_restaures

        elif choix == '15':
            afficher_aide()

        elif choix == '16':
            print("\n Merci d'avoir utilis√© le syst√®me de gestion des √©tudiants !")
            print("      Statistiques finales :")
            print(f"   ‚Ä¢ {len(etudiants)} √©tudiants g√©r√©s")
            print(f"   ‚Ä¢ Moyenne g√©n√©rale : {sum(e['moyenne'] for e in etudiants)/len(etudiants):.2f}/20")
            print("   ‚Ä¢ Au revoir ! üëã")
            break
        else:
            print("Choix invalide. Veuillez r√©essayer.")

if __name__ == "__main__":
    main() 
```
